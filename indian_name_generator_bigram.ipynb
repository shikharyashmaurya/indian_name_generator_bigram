```python
import csv

# Create an empty list to store your data
words = []

# Specify the CSV file path
csv_file = '/content/Indian_Names.csv'

# Open the CSV file and read its content into the list
with open(csv_file, 'r') as file:
    csv_reader = csv.reader(file)
    for row in csv_reader:
        words.append(row)
```

```python
words[:5]
```

```python
# Create an empty list to store the names
names_list = []

# Iterate through the original list and extract the names
for item in words[1:]:  # Start from the second element to skip the header
    names_list.append(item[1])

# Now, 'names_list' contains only the names from the original list
words=names_list
```

```python
import re

# Create an empty list to store the alphabets
alphabets_list = []

# Iterate through the words and extract alphabets using regular expressions
for word in words:
    alphabets = re.sub(r'[^a-z]', '', word)
    alphabets_list.append(alphabets)

# Now, 'alphabets_list' contains only the alphabets from the words
words=alphabets_list
```

```python
N=torch.zeros((27,27),dtype=torch.int32)
```

```python
chars=sorted(list(set(''.join(words))))
chars
```

```python
chars=sorted(list(set(''.join(words)))) #find sorted list of char used
stoi={s:i+1 for i,s in enumerate(chars)} #make embedding
stoi['.']=0
itos={i:s for s,i in stoi.items()}
```

```python
for w in words:
  chs=['.']+list(w)+['.']     #start and end specifier
  for ch1,ch2 in zip(chs,chs[1:]):  #zip make pairs
    ix1=stoi[ch1]
    ix2=stoi[ch2]
    N[ix1,ix2]+=1
```

```python
P=(N+1).float()
```

```python
P/=P.sum(1,keepdim=True)
P.shape
```

```python
g=torch.Generator().manual_seed(247)
for i in range(5):
  out=[]
  ix=0
  while True:

    p=P[ix]

    # p=N[ix].float()
    # p=p/p.sum()

    # # p=torch.ones(27)/27.0  # if all uniform values are given

    ix=torch.multinomial(p,num_samples=1,replacement=True,generator=g).item()
    out.append(itos[ix])
    if(ix==0):
      break
  print(''.join(out[:-1]))
```

```python
log_likelihood=0.0
n=0

# for w in ['yash']:
for w in words:
  chs=['.']+list(w)+['.']     #start and end specifier
  for ch1,ch2 in zip(chs,chs[1:]):  #zip make pairs
    ix1=stoi[ch1]
    ix2=stoi[ch2]
    prob=P[ix1,ix2]
    logprob=torch.log(prob)
    log_likelihood+=logprob
    n+=1

print(f'{log_likelihood=}')
nll=-log_likelihood
print(f'{nll=}')
print(f'{nll/n}')
```
